- Write a loop that extracts the lyrics of all songs
- Collect all strings in a list

```python
import os
import re

lyrics_list = []
artist_list = []
for filename in os.listdir('data/song_files'):
    #print(filename)
    html_string = open('data/song_files/' + filename).read()
    
    # convert to BeautifulSoup object
    soup_object = BeautifulSoup(html_string)
    lyrics = soup_object.find('pre', attrs={'class': 'lyric-body'}).text
    lyrics_list.append(lyrics)
    
    # get a list of the artists
    artist_pattern = '.+?(?=_)'
    artist_name = re.findall(artist_pattern, filename)
    artist_list.append(artist_name) # regex to get only the artist name

        
    
```

```python
from itertools import chain
artist_list = list(chain(*artist_list))
```

### Preprocessing with spacy

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_md')
```

```python
# function from the lecture
def clean_text(corpus, model):
    """preprocess a string (tokens, stopwords, lowercase, lemma & stemming) returns the cleaned result
        params: review - a string
                model - a spacy model
                
        returns: list of cleaned strings
    """
    
    new_doc = []
    doc = model(corpus)
    for word in doc:
        if not word.is_stop and word.is_alpha:
            new_doc.append(word.lemma_.lower())
            
    return new_doc
```

```python
# test out with one songtext
first_song_clean = clean_text(lyrics_list[0], nlp)
first_song_clean
```

```python
### apply with a for-loop:
song_spacy = []
for corpus in lyrics_list:
    song_cleaned = clean_text(corpus, nlp)
    song_spacy.append(song_cleaned)
```

### Create and save dataframe

```python
import pandas as pd
df = pd.DataFrame(list(zip(artist_list, lyrics_list, song_spacy)), 
               columns =['artist', 'lyrics', 'song_strings_lists']) 
dictionary = {'\+':' '}
df.replace(dictionary, regex=True, inplace=True)
df
```

```python
df.to_csv('data/df/artist_song.csv', index_label=False)
```
